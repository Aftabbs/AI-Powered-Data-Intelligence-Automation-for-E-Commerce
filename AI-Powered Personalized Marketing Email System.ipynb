### DeepSeek R1(**Experimental**)

```python
from groq import Groq
import random
import os

api_key=os.environ.get('GROQ_API_KEY')
client = Groq(api_key=api_key)

def generate_personalized_email(customer_name, last_purchase, preferred_category, discount_offer):
    """
    Generates a personalized marketing email using Groq LLM.
    """
    prompt = f"""
    Generate a personalized marketing email for a customer:

    Customer Name: {customer_name}
    Last Purchase: {last_purchase}
    Preferred Category: {preferred_category}
    Special Discount Offer: {discount_offer}

    The email should be engaging, include a compelling subject line, and have a strong call-to-action.
    """

    completion = client.chat.completions.create(
        model="deepseek-r1-distill-llama-70b",
        messages=[
            {"role": "user", "content": prompt}
        ],
        temperature=0.2,
        max_completion_tokens=1000,
        top_p=0.9,
        stream=False
    )

    return completion.choices[0].message.content

customer_name = "John Doe"
last_purchase = "Wireless Headphones"
preferred_category = "Electronics"
discount_offer = "20% off on your next purchase!"

email_content = generate_personalized_email(customer_name, last_purchase, preferred_category, discount_offer)
print(email_content)
```

### Prompting With Few Shots

```python
from groq import Groq
import pandas as pd

# Create a sample DataFrame
customers = [
    {"customer_name": "John Doe", "last_purchase": "Wireless Headphones", "preferred_category": "Electronics", "discount_offer": "20% off on your next purchase!"},
    {"customer_name": "Jane Smith", "last_purchase": "Running Shoes", "preferred_category": "Sportswear", "discount_offer": "15% off on sports gear!"}
]

customer_df = pd.DataFrame(customers)

few_shot_examples = """
### Example 1:
Customer: Mark Taylor
Last Purchase: Smartwatch
Preferred Category: Wearables
Discount Offer: 10% off on all smart gadgets!

Email:
Subject: Mark, Your Exclusive 10% Discount on Smart Gadgets Awaits!

Hello Mark,

We noticed you loved your new Smartwatch from ABC Corp Ltd. To make your next purchase even better, we're offering **10% off on all smart gadgets!**

Visit [ABCcorp.com](https://www.abccorp.com) and use code **SMART10** at checkout. Offer valid for the next 48 hours!

For any assistance, call us at **12345684** or DM us on **@abccorp4all**.

Happy Shopping,
ABC Corp Ltd Team

---

### Example 2:
Customer: Sarah Lee
Last Purchase: Designer Handbag
Preferred Category: Fashion
Discount Offer: 25% off on your next fashion item!

Email:
Subject: Sarah, Enjoy 25% Off on Your Next Fashion Purchase!

Hi Sarah,

You have a great sense of style! Since you loved your Designer Handbag, we’re giving you **25% off your next fashion buy** at ABC Corp Ltd.

Use code **FASHION25** at checkout on [ABCcorp.com](https://www.abccorp.com). Offer ends soon!

Need help? Call **12345684** or reach us on **@abccorp4all**.

Stay Stylish,
ABC Corp Ltd Team
"""

def generate_email(customer):
    prompt = f"""
    {few_shot_examples}
    ---
    Customer: {customer['customer_name']}
    Last Purchase: {customer['last_purchase']}
    Preferred Category: {customer['preferred_category']}
    Discount Offer: {customer['discount_offer']}

    Email:
    """

    completion = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[
            {"role": "user", "content": prompt}
        ],
        temperature=0.6,
        max_completion_tokens=700,
        top_p=0.95,
    )

    return completion.choices[0].message.content.strip()


customer_df['personalized_email'] = customer_df.apply(generate_email, axis=1)

print(customer_df[['customer_name', 'personalized_email']])
```

### Build a Small Email Agent Using Phi-Data

```python
from phi.agent import Agent
from phi.tools.email import EmailTools
from phi.model.groq import Groq
import os

receiver_email = "abc@gmail.com"
sender_email = "xyz@gmail.com"
sender_name = "abc"
sender_passkey = "123@abc"


agent = Agent(model=Groq(id="llama-3.3-70b-versatile",api_key=api_key),tools=[
        EmailTools(
            receiver_email=receiver_email,
            sender_email=sender_email,
            sender_name=sender_name,
            sender_passkey=sender_passkey,
        ),instructions=[
          "You are an email assistant mainly focused on marketing.",
          "Use the EmailTools to send emails.",
          "Provide concise and clear responses."
      ],
      markdown=True
    ])

customer_name = "John Doe"
last_purchase = "Wireless Headphones"
preferred_category = "Electronics"
discount_offer = "20% off on your next purchase!"
prompt = f"""
    Generate a personalized marketing email for a customer:

    Customer Name: {customer_name}
    Last Purchase: {last_purchase}
    Preferred Category: {preferred_category}
    Special Discount Offer: {discount_offer}

    The email should be engaging, include a compelling subject line, and have a strong call-to-action.
    """
agent.print_response(f"{prompt} to {receiver_email}",stream=True,debug=True)
```
